```python


dataset = fetch_olivetti_faces()
images = dataset.images

# Create a figure with 10 rows and 10 columns
fig,ax = plt.subplots(10, 10, figsize=(10, 10))

# Display each image in its own cell
for i in range(10):
    for j in range(10):
        ax[i, j].imshow(images[i * 10 + j], cmap='gray')
        ax[i, j].axis('off')

plt.show()
```
